Nominal data visualization

```python
import matplotlib.pyplot as plt

cities = ['Nanjing', 'Shanghai', 'Beijing', 'Guangzhou', 'Shenzhen']
population = [9.3, 24.87, 21.89, 18.68, 17.66]

plt.figure(figsize=(10, 6))
plt.bar(cities, population, color='skyblue')

plt.xlabel('City')
plt.ylabel('Population')
plt.title('City Population')

plt.show()
```

Ordinal data visualization:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset with ordinal data
categories = ['Low', 'Medium', 'High', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High', 'Medium']
values = [10, 20, 30, 25, 15, 35, 22, 12, 32, 28]

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=categories, y=values, order=['Low', 'Medium', 'High'])

# Add title and labels
plt.title('Box Plot of Ordinal Data')
plt.xlabel('Category')
plt.ylabel('Values')

# Show the plot
plt.show()
```

Histograms for the above:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed
np.random.seed(0)

# Generate data
# Category A: Normal distribution
data_a = np.random.normal(loc=50, scale=5, size=200)

# Category B: Positively skewed distribution
data_b = np.random.exponential(scale=10, size=200) + 30

# Category C: Bimodal distribution
data_c1 = np.random.normal(loc=40, scale=3, size=100)
data_c2 = np.random.normal(loc=60, scale=3, size=100)
data_c = np.concatenate([data_c1, data_c2])

# Create DataFrame
data = pd.DataFrame({
    'Category': ['A'] * len(data_a) + ['B'] * len(data_b) + ['C'] * len(data_c),
    'Value': np.concatenate([data_a, data_b, data_c])
})

# Set style
sns.set_style('whitegrid')

# Define order
order = ['A', 'B', 'C']

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 12))

for i, category in enumerate(order):
    subset = data[data['Category'] == category]['Value']
    
    # Boxplot with Swarm Plot (Left Column)
    sns.boxplot(x=subset, ax=axes[i, 0], color='lightblue', showfliers=True)
    #sns.swarmplot(x=subset, ax=axes[i, 0], color='black', size=3)
    axes[i, 0].set_title(f'Category {category} - Boxplot')
    axes[i, 0].set_xlabel('Value')
    axes[i, 0].set_ylabel('')
    
    # Distribution Plot (Right Column)
    sns.histplot(subset, kde=True, ax=axes[i, 1], color='lightcoral', bins=20)
    axes[i, 1].set_title(f'Category {category} - Distribution')
    axes[i, 1].set_xlabel('Value')
    axes[i, 1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

WIth negative skeweness

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# Set random seed
np.random.seed(0)

# Generate data
# Category A: Normal distribution
data_a = np.random.normal(loc=50, scale=5, size=200)

# Category B: Positively skewed distribution
data_b = np.random.exponential(scale=10, size=200) + 30

# Category C: Bimodal distribution
data_c1 = np.random.normal(loc=40, scale=3, size=100)
data_c2 = np.random.normal(loc=60, scale=3, size=100)
data_c = np.concatenate([data_c1, data_c2])

# Category D: Negatively skewed distribution
data_d = skewnorm.rvs(a=-5, loc=50, scale=5, size=200)

# Create DataFrame
data = pd.DataFrame({
    'Category': (['A'] * len(data_a)) + 
                (['B'] * len(data_b)) + 
                (['C'] * len(data_c)) + 
                (['D'] * len(data_d)),
    'Value': np.concatenate([data_a, data_b, data_c, data_d])
})

# Set style
sns.set_style('whitegrid')

# Define order
order = ['A', 'B', 'C', 'D']

# Create subplots
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(14, 16))

for i, category in enumerate(order):
    subset = data[data['Category'] == category]['Value']
    
    # Boxplot with Swarm Plot (Left Column)
    sns.boxplot(x=subset, ax=axes[i, 0], color='lightblue', showfliers=True)
    axes[i, 0].set_title(f'Category {category} - Boxplot')
    axes[i, 0].set_xlabel('Value')
    axes[i, 0].set_ylabel('')
    
    # Distribution Plot (Right Column)
    sns.histplot(subset, kde=True, ax=axes[i, 1], color='lightcoral', bins=20)
    axes[i, 1].set_title(f'Category {category} - Distribution')
    axes[i, 1].set_xlabel('Value')
    axes[i, 1].set_ylabel('Frequency')
    
    # Adjust x-limits for consistency
    axes[i, 0].set_xlim(data['Value'].min() - 5, data['Value'].max() + 5)
    axes[i, 1].set_xlim(data['Value'].min() - 5, data['Value'].max() + 5)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

Boxplot using only matplotlib

```python

import matplotlib.pyplot as plt

# Dataset for Group A with an outlier
data_group_a = [12, 15, 14, 10, 13, 17, 15, 14, 13, 16, 15, 14, 13, 35]

# Dataset for Group B with an outlier
data_group_b = [22, 25, 24, 20, 23, 27, 25, 24, 23, 26, 25, 24, 23, 2]

# Data for plotting
data = [data_group_a, data_group_b]
labels = ['Group A', 'Group B']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create the boxplot
ax.boxplot(data, labels=labels)

# Set title and labels
ax.set_title('Boxplots with Outliers using only Matplotlib')
ax.set_xlabel('Group')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
